```python
import pandas as pd
import random as rd

from lshash.lshash import LSHash
```

```python
#è¯»å–æ–‡ä»¶
df = pd.read_csv('../reco_data/trade.csv')

#è£å‰ªï¼Œå¹¶æ ¹æ®vipnoã€plunoåˆ†ç»„ï¼Œå¯¹amtæ±‚å’Œ
df1 = df.groupby([df.vipno, df.pluno])[['amt']].sum()
#è¿™é‡Œå°†dataè½¬æ¢ä¸ºä»¥vipnoä¸ºè¡Œï¼Œplunoä¸ºåˆ—ï¼Œä¾¿äºŽä¹‹åŽå¤„ç†
df2 = df1.unstack(0).fillna(0).round().transpose().loc['amt']
#å°†vipnoè½¬æ¢ä¸ºåˆ—ï¼Œå¯¹ä¸å­˜åœ¨çš„å€¼å¡«0ï¼Œæ•°å€¼å››èˆäº”å…¥
data = df2.transpose()
#åŽ»æŽ‰è¡Œå¤´å’Œåˆ—å¤´
del data.index.name
del data.transpose().index.name

data
```

```python
#å°†dataframeè½¬æ¢ä¸ºçº¯æ•°å­—çŸ©é˜µ
data_matrix = data.as_matrix()
col = data.shape[1]
row = data.shape[0]

for size in [0.01, 0.05, 0.1, 0.2, 0.3, 0.5]:
    hash_size = int(col * size)
    lsh = LSHash(hash_size, input_dim = row )
    
    #å°†çŸ©é˜µå¯¼å…¥ï¼Œç”Ÿæˆå“ˆå¸Œè¡¨
    for  col_index in range(col):
        #å°†çŸ©é˜µçš„æ¯ä¸€åˆ—å¯¼å…¥å“ˆå¸Œè¡¨ï¼Œå°†vipnoçš„å€¼ä½œä¸ºextra_data
        lsh.index(data_matrix[:, col_index], extra_data = data.columns[col_index])
    else:
        #éšæœºå–åˆ—
        vipno_pos = 14
        for k in [1, 2, 3, 4, 5] :
            #å¾—åˆ°hash_tableçš„æŸ¥è¯¢ç»“æžœ
            hash_table =  lsh.query(data_matrix[:, vipno_pos], num_results= k + 1, distance_func="euclidean")
            print("hash_size:" + str(hash_size) +
                  ", k:" + str(k) +
                  ", vipno:" +  str(data.columns[vipno_pos]) +
                  ", result:")
            result = []
            for res in hash_table:
                result.append(res[0][1])
            else:
                print(result[1:])
```
